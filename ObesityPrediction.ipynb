**Removing irrelevant columns and records**

```python
#this column is being removed after looking its countplot and classes
df.drop(columns = 'Transport', inplace = True)
```

```python
#removing record as there is only one record with this value
df = df[df['Alcohol'] != 'Always']
```

```python
df['Alcohol'].unique()
```

**Label encoding target column**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Health'] = le.fit_transform(df['Health'])
df
```

##**Creating Column Transformers**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop(columns = 'Health'), df['Health'], test_size = 0.2, random_state = 3)
print(x_train)
print(y_train)
```

```python
df
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
import xgboost

#one hot encoder column transfomer
trf1 = ColumnTransformer([('one_hot_encoder', OneHotEncoder(sparse_output = False, handle_unknown = 'ignore', drop = 'first'), [0, 4, 5, 6])],
                         remainder = 'passthrough')

#standard scaler column transformer
trf2 = ColumnTransformer([('scaler', StandardScaler(), slice(0, 8))])

#XGBoost model
trf3 = xgboost.XGBClassifier(max_depth = 9, learning_rate = 0.3)
```

**Note - I have already tried and tested various classification models like SVM, RandomForest with hyper parameter tuning. Among them, XGBoost gave the best results hence the particular model is selected**

##**Creating Pipeline**

```python
from sklearn.pipeline import Pipeline
pipe = Pipeline([
    ('OneHot', trf1),
    ('Scaler', trf2),
    ('XGBClassifier', trf3)
])
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
```

**Checking Accuracy and Recall (Recall is necessary because cost of false negative is higher)**

```python
from sklearn.metrics import accuracy_score, recall_score
accuracy_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred, average = 'macro')
```

##**Exporting Pipeline**

```python
import pickle
pickle.dump(pipe, open('obesity_pickle.pkl', 'wb'))
```
